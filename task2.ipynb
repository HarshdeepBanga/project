```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, GlobalAveragePooling1D

# Dataset sheet2
df = pd.read_csv("Sheet_2.csv", encoding='latin-1')

print("Columns:", df.columns)
print(df.head())

# ---------------- INPUT & OUTPUT ----------------
texts = df['resume_text'].astype(str)
labels = df['class']

# Encoding
encoder = LabelEncoder()
labels = encoder.fit_transform(labels)   # converts text → numbers

# ---------------- TOKENIZATION ----------------
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

# ---------------- PADDING ----------------
max_length = 200
X = pad_sequences(sequences, maxlen=max_length)

# ---------------- SPLIT ----------------
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.2, random_state=42
)

# ---------------- MODEL ----------------
model = Sequential([
    Embedding(input_dim=10000, output_dim=64, input_length=max_length),
    GlobalAveragePooling1D(),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ---------------- SUMMARY ----------------
print("\nModel Summary:")
model.summary()

# ---------------- TRAIN ----------------
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# ---------------- EVALUATE ----------------
loss, acc = model.evaluate(X_test, y_test)
print("\nAccuracy:", acc)
```

USING DATASET Sheet1.csv --> Accuracy: 87.5 --> simpler patterns due to which accuracy comes out higher than above dataset


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, GlobalAveragePooling1D

# Dataset sheet2
df = pd.read_csv("Sheet_1.csv", encoding='latin-1')

print("Columns:", df.columns)
print(df.head())

# ---------------- INPUT & OUTPUT ----------------
texts = df['response_text'].astype(str)
labels = df['class']

# Encoding
encoder = LabelEncoder()
labels = encoder.fit_transform(labels)   # converts text → numbers

# ---------------- TOKENIZATION ----------------
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

# ---------------- PADDING ----------------
max_length = 200
X = pad_sequences(sequences, maxlen=max_length)

# ---------------- SPLIT ----------------
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.2, random_state=42
)

# ---------------- MODEL ----------------
model = Sequential([
    Embedding(input_dim=10000, output_dim=64, input_length=max_length),
    GlobalAveragePooling1D(),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ---------------- SUMMARY ----------------
print("\nModel Summary:")
model.summary()

# ---------------- TRAIN ----------------
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# ---------------- EVALUATE ----------------
loss, acc = model.evaluate(X_test, y_test)
print("\nAccuracy:", acc)
```

**Why use Embedding layer instead of raw numbers?**

Because:

-> Raw numbers only represent word indices and do not capture meaning

-> Embedding layer converts words into dense vectors representing semantic relationships

-> This allows the model to understand context and improves learning performance
